```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tkinter import Tk
from tkinter.filedialog import askopenfilename

def detect_dark_spots(image_path, threshold=100):
    # Baca gambar
    image = cv2.imread(image_path)
    
    # Konversi ke skala abu-abu
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Ambil nilai pixel yang lebih kecil dari threshold
    dark_spots = np.where(gray < threshold)
    
    # Tampilkan titik-titik gelap pada gambar
    for y, x in zip(*dark_spots):
        cv2.circle(image, (x, y), 3, (0, 0, 255), -1)  # Gambar lingkaran pada titik gelap
    
    # Tampilkan gambar dengan titik gelap
    plt.figure(figsize=(10,7))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title('Deteksi Titik Gelap')
    plt.show()

# Memilih gambar dari dialog
root = Tk()
root.withdraw()
image_path = askopenfilename(title="Pilih gambar")

# Panggil fungsi dengan path gambar sebagai argumen
detect_dark_spots(image_path)
```
